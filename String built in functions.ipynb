# 21 String .isspace()
# Invisible spacing characters like:
# " " → space,"\t" → tab,"\n" → newline
# returns True if all characters are whitespace, otherwise False.

```python
txt = "   s   "
x = txt.isspace()
print(x)
```

```python
data = ["Delhi", "   ", "\t", "Mumbai"]

for city in data:
    if not city.isspace():
        print(city)
```
